```python
import sys

import io
import pandas
import requests

#io and requests allow direct reference to csv via a link

'''
Task 1: Loading the Data:

    Load the Titanic dataset into a Pandas DataFrame.
'''

url = "https://educlaas.omnilearnlms.sambaash.com/assets/courseware/v1/9b75560db14fbe6a5e86cb55f72faf70/asset-v1:EDUCLaaS+PYD+PDDS-PYD-0424-24Jul2024+type@asset+block/titanic.csv"
serverRequest = requests.get(url).content

#csvFile already counts as a DataFrame on read
csvFile = pandas.read_csv(io.StringIO(serverRequest.decode('utf-8')))
```

```python
'''
Task 2: Initial Exploration:

    Display the first 5 rows of the dataset to get an overview.
    Provide basic information about the dataset using the info() method.
'''
# isnull checks if any columns have null values
print(csvFile.head(), "\n\n")
csvFile.info()
```

```python
'''
Task 3: Handling Missing Values:

    Identify and count the number of missing values in each column.
    Decide on an appropriate strategy for handling missing values in at least two columns and implement it.

'''
print(csvFile.isnull().any(), "\n\n")
print('Nulls in Age column: ', csvFile['Age'].isna().sum(),'\n Nulls in Cabin column: ', csvFile['Cabin'].isna().sum(), '\n Nulls in Embarked column: ', csvFile['Embarked'].isna().sum())
csvFile[['Cabin', 'Embarked']] = csvFile[['Cabin', 'Embarked']].fillna('Unknown')

print(csvFile.head(50))
```

```python
'''
Task 4: Descriptive Statistics:

    Generate descriptive statistics for the numerical columns using the describe() method.

'''
pandas.set_option('display.float_format', lambda x: '%.3f' % x)
print(csvFile.describe())

'''
-> there were 891 people in the record
-> 38% survived
-> a lot of them were in Pclass 1
-> a lot of them were at most 40 years old
-> half of the people had siblings
-> 75% of the people paid less than 31 dollars to get on board
'''
```

```python
'''
Task 5: Unique Values:

    List unique values in the 'Pclass' column.
    Count the occurrences of each unique value in the 'Embarked' column.

'''
print(csvFile.Pclass.unique().tolist())
print("\n")
print(csvFile.Embarked.value_counts())
```

```python
'''
Task 6: Data Manipulation:

    Create a new column named 'FamilySize' that represents the sum of 'SibSp' and 'Parch'.
    Drop the 'Cabin' column from the dataset.

'''
csvFile['FamilySize'] = csvFile['Parch'] + csvFile['SibSp']
csvFile.drop(columns = ['Cabin'])
csvFile.head()
```

```python
'''
Task 7: Filtering and Sorting:

    Create a new DataFrame containing only passengers with an age greater than 30.
    Sort the DataFrame by 'Fare' in descending order.

'''
csvFile_30 = csvFile[csvFile['Age'] > 30]
csvFile_30.sort_values(by='Fare', ascending=False)
csvFile_30.head(50)
```

```python
'''
Task 8: Data Analysis:

    Calculate the survival rate for different passenger classes ('Pclass').
    Identify and display the passenger with the highest 'Fare'.

'''
#input will be the 3 unique values of Pclass
def survival_rate(inputArray):
  for i in range(len(inputArray)):
    #count number of 1s within each unique value
    #then divide it by total of that unique value = %
    csvFileQueried = csvFile[csvFile['Pclass'] == i+1 ]
    csvQueriedInt = int(len(csvFileQueried))
    csvQueriedSurvival = int(len(csvFileQueried[csvFileQueried['Survived'] == 1]))
#    print(i, "/t", csvQueriedInt, "/t", csvQueriedSurvival)
    csvSurvivalRate = csvQueriedSurvival / csvQueriedInt
    print("The survival rate for Pclass ", inputArray[i], " is ", csvQueriedSurvival, " / ", csvQueriedInt, " or ", round(csvSurvivalRate * 100), "%")

pClassValues = sorted( csvFile['Pclass'].unique() )
survival_rate(pClassValues)
#csvFile.dtypes


csvFileFare = csvFile[csvFile.Fare == csvFile.Fare.max()]
print("The following have the highest fares in the Titanic: \n", csvFileFare.get('Name').tolist())
```
