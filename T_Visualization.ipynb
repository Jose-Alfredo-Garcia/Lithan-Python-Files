# matplotlib - Line Plot

    Displays data points connected by straight lines, useful for showing trends over time.
    Usage: Visualizes data points connected by straight lines.
    Scenario: Ideal for showing trends over time, such as stock prices, temperature changes, or sales data.

````python
import matplotlib.pyplot as plt

# Sample data
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May']
sales = [100, 150, 50, 250, 300]

# Creating a line plot
plt.plot(months, sales)
plt.title('Monthly Sales')
plt.xlabel('Months')
plt.ylabel('Sales')
plt.grid(False)
plt.show()
````

Key Components:

    Data Points: Represented by markers (e.g., circles) on the plot, indicating specific values.
    Line: Connects the data points, showing the trend or change over time.
    Axes Labels: Indicate what the X-axis and Y-axis represent (e.g., time, measurement units).
    Title: Describes the overall subject of the plot.
    Grid Lines: Optional but can enhance readability by making it easier to match data points with their corresponding values on the axes.

````python
import matplotlib.pyplot as plt

# Sample data: months and corresponding sales
months = ['January', 'February', 'March', 'April', 'May']
sales = [2500, 3000, 2000, 3500, 4000]

# Create a line plot
plt.figure(figsize=(8, 5))  # Set the size of the plot
plt.plot(months, sales, marker='o', linestyle='-', color='b')  # Plot data with blue line and circle markers

# Add title and labels
plt.title('Monthly Sales', fontsize=14)  # Title of the plot
plt.xlabel('Months', fontsize=12)  # X-axis label
plt.ylabel('Sales (in dollars)', fontsize=12)  # Y-axis label

# Show grid for better readability
plt.grid(True)

# Display the plot
plt.show()
````

# matplotlib - Bar Plot

    Represents data with rectangular bars with lengths proportional to the values they represent.
    Usage: Represents categorical data with rectangular bars.
    Scenario: Useful for comparing different categories, such as sales by product type or population by country.

````python
# Sample data
products = ['Product A', 'Product B', 'Product C']
sales = [300, 450, 500]

# Creating a bar plot
plt.bar(products, sales)
plt.title('Sales by Product')
plt.xlabel('Products')
plt.ylabel('Sales')
plt.show()
````

Key Components:

    Bars: Represent the categories being compared. The height or length corresponds to the value.
    Categories: Labels for each bar, displayed along the X-axis (for vertical bars) or Y-axis (for horizontal bars).
    Values: The numerical values represented by the length or height of the bars.
    Title and Labels: Provide context and clarify what the plot represents.
    Colors and Edges: Bars can be colored differently for better distinction and aesthetic appeal, and edges can be outlined for emphasis.

````python
import matplotlib.pyplot as plt

# Sample data: products and their sales
products = ['Product A', 'Product B', 'Product C', 'Product D']
sales = [1500, 2300, 3200, 4100]

# Create a bar plot
plt.figure(figsize=(8, 5))  # Set the size of the plot
plt.bar(products, sales, color='orange')  # Create bars with orange color

# Add title and labels
plt.title('Sales by Product', fontsize=14)  # Title of the plot
plt.xlabel('Products', fontsize=12)  # X-axis label
plt.ylabel('Sales (in dollars)', fontsize=12)  # Y-axis label

# Show the plot
plt.show()
````

# matplotlib - Histogram

    A graphical representation that organizes a group of data points into user-specified ranges. Each bar represents the frequency of data points within a range.
    Usage: Displays the distribution of a dataset.
    Scenario: Used to understand the frequency distribution of continuous data, such as exam scores or age distributions.

````python
# Sample data
ages = [22, 25, 27, 25, 30, 32, 35, 35, 40, 42, 45]

# Creating a histogram
plt.hist(ages, bins=5)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
````

Key Components:

    Data: Numerical values that you want to analyze.
    Bins: Intervals that define the range of data points. The number of bins can significantly affect the appearance and interpretation of the histogram.
    Frequency: The count of data points within each bin.

````python
import matplotlib.pyplot as plt

# Sample data: ages of a group of people
ages = [22, 25, 29, 30, 32, 35, 37, 40, 45, 50, 55, 60, 65, 70, 75, 80]

# Create a histogram
plt.figure(figsize=(8, 5))  # Set the size of the plot
plt.hist(ages, bins=8, color='purple', edgecolor='black')  # Create histogram with 8 bins and purple color

# Add title and labels
plt.title('Age Distribution', fontsize=14)  # Title of the plot
plt.xlabel('Age', fontsize=12)  # X-axis label
plt.ylabel('Frequency', fontsize=12)  # Y-axis label

# Show the plot
plt.show()

````

# matplotlib - Scatter Plot

    Displays values for two variables for a set of data points, often used to determine if there is a correlation between the two.
    Usage: Shows the relationship between two numerical variables.
    Scenario: Helpful for identifying correlations or patterns, such as height vs. weight or income vs. expenditure.

````python
# Sample data
height = [150, 160, 170, 180, 190]
weight = [50, 60, 70, 80, 90]

# Creating a scatter plot
plt.scatter(height, weight)
plt.title('Height vs. Weight')
plt.xlabel('Height (cm)')
plt.ylabel('Weight (kg)')
plt.show()
````

Key Components:

    Data Points: Represented by dots or other markers, each point corresponds to one observation in the dataset.
    X-axis and Y-axis: Represent the two variables being compared. The position of a point on these axes indicates its value for the respective variables.
    Title and Labels: Provide context and explain what the plot represents.
    Grid Lines: Optional but can enhance readability by helping to align points with their corresponding values on the axes.
    Markers: Can vary in shape, size, and color to distinguish between different datasets or highlight specific data points.

````python
import matplotlib.pyplot as plt

# Sample data: age and income
ages = [25, 30, 35, 40, 45, 50, 55, 60]
incomes = [30000, 35000, 50000, 60000, 65000, 70000, 75000, 80000]

# Create a scatter plot
plt.figure(figsize=(8, 5))  # Set the size of the plot
plt.scatter(ages, incomes, color='blue', marker='x', s=100)  # Plot data points with blue 'x' markers

# Add title and labels
plt.title('Income vs. Age', fontsize=14)  # Title of the plot
plt.xlabel('Age (years)', fontsize=12)  # X-axis label
plt.ylabel('Income (in dollars)', fontsize=12)  # Y-axis label

# Display grid lines for better readability
plt.grid(True)

# Show the plot
plt.show()
````

# matplotlib - Pie Chart

    A circular chart divided into sectors, each representing a proportion of the total. The size of each sector is proportional to the quantity it represents.
    Usage: Displays proportions of a whole.
    Scenario: Suitable for showing percentage or proportional data, like market share or survey results.

````python
# Sample data
labels = ['Apples', 'Bananas', 'Cherries', 'Dates']
sizes = [15, 30, 45, 10]

# Creating a pie chart
plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.title('Fruit Distribution')
plt.show()
````

Key Components:

    Data: Numerical values representing the parts of the whole.
    Labels: Descriptions for each sector, indicating what each part represents.
    Autopct: A parameter that formats the display of the percentage each sector represents. For example, %1.1f%% shows the percentage with one decimal place.
    Startangle: The starting angle for the first sector, which can help in positioning the pie chart for better readability.

````python
import matplotlib.pyplot as plt

# Sample data: market share of different companies
companies = ['Company A', 'Company B', 'Company C', 'Company D']
market_share = [30, 25, 20, 25]  # Market share percentages

# Create a pie chart
plt.figure(figsize=(8, 8))  # Set the size of the plot
plt.pie(market_share, labels=companies, autopct='%1.1f%%', startangle=240, colors=['gold', 'lightblue', 'lightgreen', 'lightcoral'])  # Create pie chart with percentage display and custom colors

# Add title
plt.title('Market Share by Company', fontsize=14)  # Title of the plot

# Display the plot
plt.show()
````

# matplotlib - Box Plot

    A visual representation of the distribution of a dataset. It shows the minimum, first quartile (Q1), median (Q2), third quartile (Q3), and maximum. The "whiskers" indicate variability outside the upper and lower quartiles, and any points outside the whiskers are considered outliers.
    Usage: Summarizes data through their quartiles, highlighting the median and outliers.
    Scenario: Useful for comparing distributions and detecting outliers.

````python
# Sample data
data = [22, 25, 27, 25, 30, 32, 35, 35, 40, 42, 45]

# Creating a box plot
plt.boxplot(data)
plt.title('Box Plot of Age Data')
plt.xlabel('Age')
plt.ylabel('Values')
plt.show()
````

Key Components:

    Box: Represents the interquartile range (IQR), which contains the middle 50% of the data.
    Whiskers: Extend to the smallest and largest values within 1.5 times the IQR from the lower and upper quartiles.
    Median Line: Inside the box, representing the median value of the dataset.
    Outliers: Data points outside the whiskers are considered outliers and are often marked with dots or asterisks.

   Boxplot.png

````python
import matplotlib.pyplot as plt

# Sample data: scores of students in three different subjects
math_scores = [85, 90, 78, 92, 88, 76, 95, 89, 77, 84]
science_scores = [80, 85, 75, 82, 86, 79, 91, 87, 83, 81]
english_scores = [78, 82, 88, 77, 84, 80, 85, 89, 91, 83]

# Combine data into a list
data = [math_scores, science_scores, english_scores]

# Create a box plot
plt.figure(figsize=(8, 5))  # Set the size of the plot
plt.boxplot(data, labels=['Math', 'Science', 'English'], patch_artist=True)  # Create box plot with labels and colored boxes

# Add title and labels
plt.title('Student Scores by Subject', fontsize=14)  # Title of the plot
plt.xlabel('Subject', fontsize=12)  # X-axis label
plt.ylabel('Scores', fontsize=12)  # Y-axis label

# Show the plot
plt.show()
````

# Seaborn - Distplot (Deprecated in favor of histplot and kdeplot)

    distplot: Combines a histogram and a Kernel Density Estimate (KDE) plot to visualize the distribution of data. The histogram shows the frequency of data points within bins, and the KDE provides a smoothed estimate of the data's density.
    Usage: Useful for understanding the distribution of a dataset, visualizing its spread, and identifying patterns or skewness.
    Scenario: Examining the distribution of ages in a population to see how frequently different age groups occur and understanding the overall density distribution.

Key Components:

    Histogram Bars: Represent the frequency of data points within specified bins.
    KDE Curve: A smooth line that estimates the data's density, giving a clearer view of the distribution's shape.
    Bins: The number of bins in the histogram can be adjusted to change the granularity of the distribution visualization.
    Color and Edgecolor: Customize the appearance of the histogram bars for better visual distinction.

Parameters:

    kde: Boolean value to include or exclude the KDE curve (default is False). Setting kde=True adds the KDE line to the histogram.
    bins: Number of bins for the histogram, which controls the width of the bins.
    color: Color of the histogram bars.
    edgecolor: Color of the edges of the histogram bars, useful for better separation between bars.

````python
import seaborn as sns
import matplotlib.pyplot as plt

# Sample data: ages of a group of people
ages = [22, 25, 29, 30, 32, 35, 37, 40, 45, 50, 55, 60, 65, 70, 75, 80]

# Create a distribution plot
plt.figure(figsize=(8, 5))  # Set the size of the plot
#sns.histplot(ages, kde=True, bins=10, color='skyblue', edgecolor='black')  # Plot histogram with KDE
sns.distplot(ages, kde=True, bins=10, color='skyblue', hist_kws={'edgecolor': 'black'})

# Add title and labels
plt.title('Age Distribution with KDE', fontsize=14)  # Title of the plot
plt.xlabel('Age (years)', fontsize=12)  # X-axis label
plt.ylabel('Density', fontsize=12)  # Y-axis label

# Show the plot
plt.show()
````

# Seaborn - Box plot

    A graphical representation of the distribution of data that shows the median, quartiles, and potential outliers.
    Usage: Useful for comparing distributions across different categories, identifying central tendencies, and detecting outliers.
    Scenario: Comparing student scores across different subjects.
    Key Components:
        Box: Represents the interquartile range (IQR), which includes the middle 50% of the data.
        Whiskers: Extend to the smallest and largest values within 1.5 times the IQR from the quartiles.
        Median Line: A line inside the box indicating the median value.
        Outliers: Points outside the whiskers are considered outliers.

````python
import seaborn as sns
import matplotlib.pyplot as plt

# Sample data: scores in different subjects
data = {
    'Subject': ['Math', 'Math', 'Math', 'Science', 'Science', 'Science', 'English', 'English', 'English'],
    'Score': [85, 90, 78, 88, 76, 95, 84, 80, 85]
}

# Create a box plot
plt.figure(figsize=(8, 5))  # Set the size of the plot
sns.boxplot(x='Subject', y='Score', data=data, palette='Set3')  # Plot box plot with custom color palette

# Add title and labels
plt.title('Scores by Subject', fontsize=14)  # Title of the plot
plt.xlabel('Subject', fontsize=12)  # X-axis label
plt.ylabel('Score', fontsize=12)  # Y-axis label

# Show the plot
plt.show()
````

Key Components:

    Box: Represents the interquartile range (IQR), which contains the middle 50% of the data.
    Whiskers: Extend to the smallest and largest values within 1.5 times the IQR from the quartiles. They show the range of the data.
    Median Line: A line inside the box indicating the median value of the dataset.
    Outliers: Points outside the whiskers are considered outliers and are typically marked with dots or asterisks.

Parameters:

    x and y: Variables to be plotted on the X-axis and Y-axis, respectively.
    data: Data source (e.g., a DataFrame or dictionary).
    palette: Color palette for the plot, which can be customized to improve visual appeal.

````python
import seaborn as sns
import matplotlib.pyplot as plt

# Sample data: scores in different subjects
data = {
    'Subject': ['Math', 'Math', 'Math', 'Science', 'Science', 'Science', 'English', 'English', 'English'],
    'Score': [85, 90, 78, 88, 76, 95, 84, 80, 85]
}

# Create a box plot
plt.figure(figsize=(8, 5))  # Set the size of the plot
sns.boxplot(x='Subject', y='Score', data=data, hue='Subject', palette='Set2')  # Create box plot with custom color palette

# Add title and labels
plt.title('Scores by Subject', fontsize=14)  # Title of the plot
plt.xlabel('Subject', fontsize=12)  # X-axis label
plt.ylabel('Score', fontsize=12)  # Y-axis label

# Show the plot
plt.show()





````

Key Components:

    Box: Represents the interquartile range (IQR) within each category and sub-group.
    Whiskers: Extend to the smallest and largest values within 1.5 times the IQR from the quartiles for each sub-group.
    Median Line: Indicates the median value within each category and sub-group.
    Outliers: Points outside the whiskers are marked for each sub-group.
    Hue: Differentiates data within each category by another variable, adding a layer of comparison.

Parameters:

    hue: Variable used to add sub-group differentiation within each category (e.g., 'Gender').
    palette: Color palette for the plot to visually distinguish between the different sub-groups.

````python
import seaborn as sns
import matplotlib.pyplot as plt

# Sample data: scores in different subjects categorized by gender
data = {
    'Subject': ['Math', 'Math', 'Math', 'Science', 'Science', 'Science', 'English', 'English', 'English',
                'Math', 'Math', 'Math', 'Science', 'Science', 'Science', 'English', 'English', 'English'],
    'Score': [85, 90, 78, 88, 76, 95, 84, 80, 85, 80, 85, 79, 82, 87, 88, 92, 90, 85],
    'Gender': ['Male', 'Male', 'Male', 'Male', 'Male', 'Male', 'Male', 'Male', 'Male',
               'Female', 'Female', 'Female', 'Female', 'Female', 'Female', 'Female', 'Female', 'Female']
}

# Create a box plot with hue
plt.figure(figsize=(10, 6))  # Set the size of the plot
sns.boxplot(x='Subject', y='Score', hue='Gender', data=data, palette='Set2')  # Create box plot with hue

# Add title and labels
plt.title('Scores by Subject and Gender', fontsize=14)  # Title of the plot
plt.xlabel('Subject', fontsize=12)  # X-axis label
plt.ylabel('Score', fontsize=12)  # Y-axis label

# Show the plot
plt.show()
````

# Seaborn - Pair Plot

    A grid of scatter plots that displays pairwise relationships between multiple variables. Each plot in the grid shows the relationship between two variables.
    Usage: Useful for exploring the relationships and distributions of multiple variables simultaneously.
    Scenario: Analyzing the relationships between different features of flowers (e.g., sepal length, sepal width, petal length, petal width).
    Key Components:
        Scatter Plots: Each cell in the grid shows a scatter plot for a pair of variables.
        Diagonal: Often shows the distribution of the variable itself, typically with histograms or KDE plots.
        Grid: Allows for the comparison of all possible pairwise combinations of variables.

````python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Sample data: features of different flowers
data = {
    'SepalLength': [5.1, 4.9, 4.7, 4.6, 5.0, 5.4, 5.9, 6.0, 6.3, 6.1],
    'SepalWidth': [3.5, 3.0, 3.2, 3.1, 3.6, 3.9, 3.0, 3.4, 3.3, 3.0],
    'PetalLength': [1.4, 1.4, 1.3, 1.5, 1.4, 1.7, 5.1, 5.5, 6.0, 5.0],
    'PetalWidth': [0.2, 0.2, 0.2, 0.2, 0.3, 0.4, 1.8, 2.1, 2.5, 1.9]
}
df = pd.DataFrame(data)

# Create a pair plot
plt.figure(figsize=(8, 8))  # Set the size of the plot
sns.pairplot(df)  # Plot pair plot for the dataframe

# Show the plot
plt.show()
````

# Seaborn - Count Plot



    Creates a bar plot where the height of each bar represents the count of observations in each category. It is particularly useful for visualizing the frequency of different categorical values.
    Usage: Ideal for examining the distribution of categorical data, identifying the most and least common categories, and detecting patterns or imbalances.
    Scenario: Counting and comparing the number of occurrences of different categories in a dataset.

Key Components:

    Bars: Represent the count of observations for each category. The height of each bar corresponds to the number of occurrences of the respective category.
    Categories: Displayed on the X-axis, representing different groups or labels in the dataset.
    Count: The Y-axis shows the count or frequency of each category.
    Color Palette: Customizes the colors of the bars for better visual distinction. The palette parameter allows you to use predefined color palettes or specify custom colors.

Parameters:

    x: The categorical variable to plot on the X-axis.
    data: The source of the data, such as a DataFrame or dictionary.
    palette: Color palette for the plot. You can use predefined palettes (e.g., 'viridis', 'pastel') or customize the colors.

````python
import seaborn as sns
import matplotlib.pyplot as plt

# Sample data: categories and their counts
data = {
    'Category': ['A', 'B', 'A', 'C', 'B', 'A', 'C', 'C', 'B', 'A']
}

# Create a count plot
plt.figure(figsize=(8, 5))  # Set the size of the plot
sns.countplot(x='Category', data=data, palette='viridis')  # Create count plot with color palette

# Add title and labels
plt.title('Count of Observations by Category', fontsize=14)  # Title of the plot
plt.xlabel('Category', fontsize=12)  # X-axis label
plt.ylabel('Count', fontsize=12)  # Y-axis label

# Show the plot
plt.show()
````

````python
import seaborn as sns
import matplotlib.pyplot as plt

# Sample data: categories and their counts
data = {
    'Category': ['A', 'B', 'A', 'C', 'B', 'A', 'C', 'C', 'B', 'A']
}

# Create a count plot
plt.figure(figsize=(8, 5))  # Set the size of the plot
ax = sns.countplot(x='Category', data=data, palette='viridis')  # Create count plot

# Add title and labels
plt.title('Count of Observations by Category', fontsize=14)  # Title of the plot
plt.xlabel('Category', fontsize=12)  # X-axis label
plt.ylabel('Count', fontsize=12)  # Y-axis label

# Add labels to each bar
for p in ax.patches:
    height = p.get_height()  # Get the height of the bar
    ax.text(p.get_x() + p.get_width() / 2., height + 0.1,  # Position of the text
            int(height),  # Text to display
            ha='center',  # Horizontal alignment
            va='bottom',  # Vertical alignment
            fontsize=12)  # Font size

# Show the plot
plt.show()

````

# Seaborn - Joint plot
Key Components of a Joint Plot:

    Scatter Plot (or other plots like hex, kde): Shows the relationship between the two continuous variables.
    Marginal Histograms/Densities: Displays the distribution of each variable along the edges of the plot.


    sns.jointplot: Creates a joint plot combining scatter plot and marginal histograms.
    
    x='Sales' and y='Profit': Specify the variables to be plotted on the X and Y axes.

    kind='scatter': Specifies the type of plot for the bivariate distribution. Other options include 'kde' for KDE plots, 'hex' for hexbin plots, and 'reg' for regression plots.

    color='skyblue': Sets the color of the scatter plot points.
    








````python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Sample data with 'Sales' and 'Profit'
data = pd.DataFrame({
    'Category': ['Electronics', 'Furniture', 'Electronics', 'Clothing', 'Furniture', 'Clothing', 'Electronics', 'Clothing', 'Furniture', 'Electronics',
                 'Electronics', 'Furniture', 'Clothing', 'Electronics', 'Furniture', 'Clothing', 'Electronics', 'Clothing', 'Furniture', 'Electronics'],
    'Sales': [200, 150, 220, 180, 160, 190, 210, 175, 165, 225,
              240, 170, 185, 230, 155, 200, 210, 160, 190, 220],
    'Profit': [50, 30, 70, 40, 20, 50, 60, 30, 20, 70,
               80, 20, 40, 70, 10, 40, 60, 30, 20, 60]
})

# Create a joint plot
plt.figure(figsize=(10, 6))  # Set the size of the plot
sns.jointplot(x='Sales', y='Profit', data=data, kind='scatter', color='skyblue')  # Create joint plot with scatter

# Add title
plt.suptitle('Joint Plot of Sales and Profit', fontsize=16, y=1.02)  # Title of the plot

# Show the plot
plt.show()
````

# A Categorical Box Plot



```
It  is used to visualize the distribution of a continuous variable across different categories. It shows the median, quartiles, and potential outliers of the continuous variable for each category. This type of plot is useful for comparing the distribution and central tendencies of the variable across different groups.
```



Key Components:

    Box: Represents the interquartile range (IQR) where the middle 50% of the data lies. The top and bottom of the box correspond to the 75th percentile (Q3) and 25th percentile (Q1), respectively.
    Median Line: A line inside the box representing the median value of the data.
    Whiskers: Extend from the box to the minimum and maximum values within 1.5 times the IQR from the quartiles.
    Outliers: Data points that fall outside the whiskers are considered outliers and are plotted individually.

````python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Sample sales data with categories
data = pd.DataFrame({
    'Category': ['Electronics', 'Furniture', 'Electronics', 'Clothing', 'Furniture', 'Clothing', 'Electronics', 'Clothing', 'Furniture', 'Electronics',
                 'Electronics', 'Furniture', 'Clothing', 'Electronics', 'Furniture', 'Clothing', 'Electronics', 'Clothing', 'Furniture', 'Electronics'],
    'Sales': [200, 150, 220, 180, 160, 190, 210, 175, 165, 225,
              240, 170, 185, 230, 155, 200, 210, 160, 190, 220]
})

# Create a box plot
plt.figure(figsize=(10, 6))  # Set the size of the plot
sns.boxplot(x='Category', y='Sales', data=data, hue="Category", palette='Set2')  # Create box plot

# Add title and labels
plt.title('Sales Distribution by Category', fontsize=16)  # Title of the plot
plt.xlabel('Category', fontsize=14)  # X-axis label
plt.ylabel('Sales Amount', fontsize=14)  # Y-axis label

# Show the plot
plt.show()
````

# Seaborn - Heatmap
A heatmap displays data in a grid format, where each cell represents a value. The color of each cell corresponds to the value it represents, with a gradient scale indicating the range of values. Darker or more intense colors typically represent higher values, while lighter colors represent lower values.New Section

````python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Sample data
data = {
    'Product_A': [250, 300, 350, 200, 400],
    'Product_B': [400, 420, 450, 380, 410],
    'Product_C': [100, 120, 140, 160, 180],
    'Product_D': [200, 220, 240, 260, 280],
    'Product_E': [300, 330, 360, 390, 420]
}
months = ['January', 'February', 'March', 'April', 'May']

df_sales = pd.DataFrame(data, index=months)


# Create a heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(df_sales, annot=True, cmap='YlGnBu')
plt.title('Monthly Sales Heatmap')
plt.xlabel('Products')
plt.ylabel('Months')
plt.show()

````

# The small circles you see in a box plot, especially when using Seaborn, are often referred to as "outliers."

# 1. Outlier Detection: Box plots are designed to visually represent the distribution of data and highlight potential outliers. Outliers are data points that fall significantly outside the typical range of values within a dataset.

# 2. IQR Calculation: The box in a box plot represents the interquartile range (IQR), which encompasses the middle 50% of the data. The whiskers (lines extending from the box) typically extend to 1.5 times the IQR.

# 3. Outlier Representation: Data points that fall beyond the whiskers are considered potential outliers and are often plotted individually as small circles or dots.

# Reasons for Outliers:
#    - Data Entry Errors: Mistakes during data collection or recording.
#    - Natural Variation: Some data points might genuinely be extreme values within the population.
#    - Measurement Errors: Inaccuracies in the measurement process.

# Handling Outliers:
#    - Investigation: Determine if outliers are due to errors and correct them if possible.
#    - Removal (Caution): In some cases, outliers might be removed, but this should be done with careful consideration as they might hold valuable information.
#    - Transformation: Data transformation techniques can sometimes reduce the impact of outliers.

# In the context of the Titanic dataset, outliers in the "Fare" variable could represent passengers who paid significantly higher prices for their tickets, potentially indicating luxury accommodations or other special circumstances.


````python
import seaborn as sns
import matplotlib.pyplot as plt

# Calculate the IQR
Q1 = df_titanic['Fare'].quantile(0.25)
Q3 = df_titanic['Fare'].quantile(0.75)
IQR = Q3 - Q1

# Define upper and lower bounds for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Filter the DataFrame to remove outliers
data_no_outliers = df_titanic[(df_titanic['Fare'] >= lower_bound) & (df_titanic['Fare'] <= upper_bound)]

# Create a box plot of passenger fares without outliers
plt.figure(figsize=(8, 6))  # Set the figure size
sns.boxplot(x='Fare', data=data_no_outliers, color='lightblue')
plt.title('Passenger Fare Distribution (Outliers Removed)', fontsize=16)
plt.xlabel('Fare', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
````
