```python
'''
Task 1: Load the Dataset and Explore

    Download the Titanic dataset and assign it to a DataFrame.
    Display the first few rows to get an overview of the data.

'''
import sys
import io
import pandas
import requests
import seaborn
import matplotlib.pyplot as plt

url = "https://educlaas.omnilearnlms.sambaash.com/assets/courseware/v1/9b75560db14fbe6a5e86cb55f72faf70/asset-v1:EDUCLaaS+PYD+PDDS-PYD-0424-24Jul2024+type@asset+block/titanic.csv"
serverRequest = requests.get(url).content

#csvFile already counts as a DataFrame on read
csvFile = pandas.read_csv(io.StringIO(serverRequest.decode('utf-8')))
```

```python
csvFile.head(5)
```

```python
'''
Task 2: Visualize Passenger Survival

    Create a bar plot using seaborn to visualize the count of survivors and non-survivors.

'''
seaborn.countplot(x='Survived', data=csvFile, palette='viridis')
```

```python
'''
Task 3: Analyze Passenger Age Distribution

    Create a histogram using matplotlib to show the distribution of passenger ages.

'''
plt.hist(csvFile['Age'], bins=8)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
'''
Task 4: Explore Class-wise Distribution

    Use seaborn to create a count plot to display the number of passengers in each class.

'''
seaborn.countplot(x='Pclass', data=csvFile, palette='viridis')
```

```python
'''
Task 5: Investigate Fare Distribution

    Create a box plot using seaborn to visualize the distribution of passenger fares.

'''
plt.boxplot(csvFile['Fare'].round(2))
plt.show()
```

```python
'''
Task 6: Correlation Between Variables

    Generate a heatmap using seaborn to visualize the correlation matrix of the numerical variables in the dataset.

data = {'Siblings':csvFile['SibSp'], 'Survival':csvFile['Survived'], 'Fare':csvFile['Fare'], 'Age':csvFile['Age'], 'Pclass':csvFile['Pclass']}
heatData = pandas.DataFrame (data, index = csvFile['PassengerId'])
seaborn.heatmap(heatData)
'''
correlation_matrix = csvFile.select_dtypes(include=['number']).corr()
# plotting correlation heatmap
dataplot = seaborn.heatmap(correlation_matrix, cmap="YlGnBu", annot=True)
```

```python
'''
Task 7: Visualize Passenger Class and Age

    Create a scatter plot using matplotlib to show the relationship between passenger age and class.

'''
plt.xticks(sorted(csvFile['Pclass'].unique()), fontsize=12)
seaborn.scatterplot(x=csvFile['Pclass'], y=csvFile['Age'] )
```
